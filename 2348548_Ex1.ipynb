```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('panda.jpg')
cv2_imshow(img)
```

```python
#Using ImageIO
import imageio as iio

#read an image
img = iio.imread("panda.jpg")

#write it in a new format
iio.imwrite("gunda.tif", img)
```

```python
#Using MatplotLib
# Python program to read image using matplotlib
##cv2 uses BGR color format
# Matplot uses RGB Color format

#importing matploylib modules
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
#read Images
img = mpimg.imread('gunda.tif')

#output Images
plt.imshow(img)
```

```python
#Using PIL : PIL is the Python Imaging Library
#Python program to read image using PIL module

# importing PIL
#Python Imaging Library
from PIL import Image

#Read image
img = Image.open('gunda.tif')

#Output Images
#img.show(img)
display(img)

#prints mode of image
#Eg- RGB, RFBA, CMYK, BGRetc.
print(img.format)

# Prints the name of the file
print("Filename: ", img.filename)

# Prints the size as a width and height tuple (in pixels)
print("Size: ", img.size)

# Prints the width of the image (in pixels)
print("Width: ", img.width)

# Prints the height of the image (in pixels)
print("Height: ", img.height)

# Closing the image
img.close()

```

```python
#Accessing pixel values
import cv2
import numpy as np

img = cv2.imread('gunda.tif')
pixel= img[100,100]
print (pixel)
# accessing only blue pixel
blue = img[100,100,0] # if we specify 1 then we can access green pixel and 2 is for red
print (blue)
```

```python
#Splitting and Merging Image Channels
b,g,r = cv2.split(img)
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)
img1 = cv2.merge((b,g,r))
cv2_imshow(img1)
```

Converting with cv library

```python
# Save the image in different formats
cv2.imwrite('panda.jpeg', img)
cv2.imwrite('panda.png', img)
cv2.imwrite('panda.tiff', img)
```

Converting with Python Image library(PIL)

```python
img = Image.open('panda.jpg')
img.save('panda.jpeg', 'JPEG')
img.save('panda.png', 'PNG')
img.save('panda.tiff', 'TIFF')
```
