```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# Load the image
img = cv2.imread('boy.jpg', cv2.IMREAD_GRAYSCALE)
if img is None:
    print("Error: Image not loaded.")
else:
    # Define the intensity values for slicing
    low = 100
    high = 200

    # Perform gray level slicing
    sliced_img = np.zeros_like(img)
    sliced_img[(img > low) & (img < high)] = 255

    # Display the original and sliced images
    cv2_imshow(img)
    cv2_imshow(sliced_img)
```

```python
img = cv2.imread('boy.jpg', cv2.IMREAD_GRAYSCALE)

# Calculate the negative of the image
neg_img = 255 - img

# Display the original and negative images
cv2_imshow(img)
cv2_imshow(neg_img)
```

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('supercom.jpg')
cv2_imshow(img)

# Read the image
img = cv2.imread('supercom.jpg', cv2.IMREAD_GRAYSCALE)

# Perform histogram equalization
eq_img = cv2.equalizeHist(img)

# Calculate the histograms
hist_original = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_equalized = cv2.calcHist([eq_img], [0], None, [256], [0, 256])

# Plot the histograms
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(hist_original, color='b')
plt.title('Original Image Histogram')
plt.xlim([0, 256])
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.plot(hist_equalized, color='b')
plt.title('Equalized Image Histogram')
plt.xlim([0, 256])
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
