## Gaussian Filter (CUser)

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def convolve(image, kernel):
    """
    Perform convolution operation between an image and a kernel.
    """
    image_height, image_width = image.shape
    kernel_height, kernel_width = kernel.shape

    # Define the output image
    output_image = np.zeros_like(image, dtype=np.float32)

    # Perform convolution
    for i in range(image_height - kernel_height + 1):
        for j in range(image_width - kernel_width + 1):
            output_image[i, j] = np.sum(image[i:i+kernel_height, j:j+kernel_width] * kernel)

    return np.uint8(output_image)

# Load an image
image = cv2.imread('APJ.jpg', cv2.IMREAD_GRAYSCALE)

# Define the kernel for the Gaussian filter
kernel = cv2.getGaussianKernel(5, 1.0)

# Apply the Gaussian filter using the custom convolution function
gaussian_filtered_image_custom = convolve(image, kernel)

# Display the filtered image
plt.imshow(gaussian_filtered_image_custom, cmap='gray')
plt.title('Gaussian Filter-User')
plt.axis('off')
plt.show()
```

##  Laplacian Filter (User)

```python
def convolve(image, kernel):
    """
    Perform convolution operation between an image and a kernel.
    """
    image_height, image_width = image.shape
    kernel_height, kernel_width = kernel.shape

    # Calculate padding size
    pad_height = kernel_height // 2
    pad_width = kernel_width // 2

    # Pad the image
    padded_image = np.pad(image, ((pad_height, pad_height), (pad_width, pad_width)), mode='constant')

    # Define the output image
    output_image = np.zeros_like(image, dtype=np.float32)

    # Perform convolution
    for i in range(image_height):
        for j in range(image_width):
            output_image[i, j] = np.sum(padded_image[i:i+kernel_height, j:j+kernel_width] * kernel)

    return np.uint8(output_image)

# Load an image
image = cv2.imread('APJ.jpg', cv2.IMREAD_GRAYSCALE)

# Define the Laplacian kernel
laplacian_kernel = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]], np.float32)

# Apply the Laplacian filter using the custom convolution function
laplacian_filtered_image_custom = convolve(image, laplacian_kernel)

# Display the filtered image
plt.imshow(laplacian_filtered_image_custom, cmap='gray')
plt.title('Laplacian Filter (Custom)')
plt.axis('off')
plt.show()
```
